#### Factors Contributing the win and loss of the team 

```python
import statsmodels.api as sm

from sklearn.preprocessing import LabelEncoder
l= LabelEncoder()

team1=l.fit_transform(match_cleaned["team1"])
team2=l.fit_transform(match_cleaned["team2"])
toss_winner=l.fit_transform(match_cleaned["toss_winner"])
toss_decision=l.fit_transform(match_cleaned["toss_decision"])
winner=l.fit_transform(match_cleaned["winner"])
player_of_match=l.fit_transform(match_cleaned["player_of_match"])
venue=l.fit_transform(match_cleaned["venue"])
result=l.fit_transform(match_cleaned["result"])

match_cleaned_new = match_cleaned.drop(['team1','team2','toss_winner','toss_decision','winner',
                             'player_of_match','player_of_match','venue','result'],axis=1)

match_cleaned_new['team1']=team1
match_cleaned_new['team2']=team2
match_cleaned_new['toss_winner']=toss_winner
match_cleaned_new['toss_decision']=toss_decision
match_cleaned_new['winner']=winner
match_cleaned_new['player_of_match']=player_of_match
match_cleaned_new['venue']=venue
match_cleaned_new['result']=result

match_cleaned_new.head(3)

x= match_cleaned_new.drop(['result'],axis=1)
y=match_cleaned_new["result"]
X_1 = sm.add_constant(x)
#Fitting sm.OLS model
model = sm.OLS(y,X_1).fit()
model.pvalues.round(2)
```

##### ----So win_by_runs & win_by_wickets are factors contributing the win and loss of the team
